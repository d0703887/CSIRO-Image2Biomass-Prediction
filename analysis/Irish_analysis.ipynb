```python
import pandas as pd
df = pd.read_csv("../data/IrishGrassClover/camera/train.csv")
df["Dry_Green_g"] = (df["Grass Dried"] * df["Herbage Mass (kg DM/ha)"] / 100 + df["Weeds Dried"] * df["Herbage Mass (kg DM/ha)"] / 100) * 1000 * (0.21 / 10000)
df["Dry_Clover_g"] = (df["Clover Dried"] * df["Herbage Mass (kg DM/ha)"] / 100) * 1000 * (0.21 / 10000)
```

```python
import sweetviz as sv
report = sv.analyze(
    df[["Dry_Green_g", "Dry_Clover_g"]],
    feat_cfg=sv.FeatureConfig()
)
report.show_notebook(w="100%", h="full")
```
